# 3. Bandingkan total penjualan dan jumlah transaksi dari masing-masing kota pada tahun 2022?

```python
df = client.query('''

SELECT
store_city,
EXTRACT(YEAR FROM date) AS YEARS,
SUM(unit_sold) AS penjualan,
SUM(store_id) AS jumlah_transaksi
FROM `ftds-hacktiv8-project.phase0_ftds_038_rmt.beverage_sales`
WHERE EXTRACT(YEAR FROM date) =2022
GROUP BY 1,2
''').to_dataframe()

df
```

# 4. Produk apa yang paling banyak diminati pada tahun 2022?

```python
df = client.query('''

SELECT
product_name,
EXTRACT(YEAR FROM date) AS YEARS,
COUNT(product_name) AS favourite_product
FROM `ftds-hacktiv8-project.phase0_ftds_038_rmt.beverage_sales`
WHERE EXTRACT(YEAR FROM date) =2022
GROUP BY 1,2
ORDER BY 3 DESC
''').to_dataframe()

df.head()
```

```python
weather_list = []

for i in df['date'].dt.month:
  if i in (3,4,5):
    weather_list.append('SPRING')
  elif i in (6,7,8):
    weather_list.append('SUMMER')
  elif i in (9,10,11):
    weather_list.append('AUTUMN')
  elif i in (12,1,2):
    weather_list.append('WINTER')

df['quarter'] = weather_list

df.head(10)
```
